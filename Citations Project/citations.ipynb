```python
csv_path = './data/aminer_papers_0.txt'
df = pd.read_json(csv_path, lines=True)
    
```

```python
df.head(100)
```

```python
id_set = set()
for row in df.itertuples(index=True, name='Pandas'):
    id_set.add(getattr(row, "id"))
```

```python
intersection_set = set()
different_set = set()

for row in df.itertuples(index=True, name='Pandas'):
    val = getattr(row, "references")
    try:
        for x in val:
            if x in id_set:
                intersection_set.add(x)
            else:
                different_set.add(x)
    except:
        continue
```
